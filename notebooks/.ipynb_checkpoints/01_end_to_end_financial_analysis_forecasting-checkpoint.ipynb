# End-to-End Financial Analysis and Forecasting

## Project Purpose

This notebook presents a full financial analysis and forecasting workflow using simulated company data. The project is designed for financial analyst, FP&A analyst, forecasting analyst, business analyst, and data analyst roles.

The analysis answers practical business questions:

1. What is driving revenue?
2. What is driving profit?
3. Which products and regions are most profitable?
4. Where is the company beating or missing budget?
5. Are discounts helping or hurting profitability?
6. Which customers are most valuable?
7. How efficiently does the company convert sales into cash?
8. Can monthly revenue be forecasted?
9. What actions should management take?

The project uses invoice-level data, monthly budget data, financial statement data, and macroeconomic indicators.

## 3. Initial Data Inspection

Before doing any financial analysis, a professional analyst should inspect the data for:

- Missing values
- Duplicate rows
- Incorrect data types
- Inconsistent categorical values
- Outliers
- Date problems

This step is important because financial decisions based on bad data can lead to bad forecasts, weak budgets, and misleading executive reporting.

```python
def inspect_dataframe(df, name):
    print(f"\n==================== {name.upper()} ====================")
    print("Shape:", df.shape)
    print("\nData types:")
    print(df.dtypes)
    print("\nMissing values:")
    print(df.isna().sum().sort_values(ascending=False).head(15))
    print("\nDuplicate rows:", df.duplicated().sum())

inspect_dataframe(transactions, "transactions")
inspect_dataframe(budget, "budget")
inspect_dataframe(financials, "financials")
inspect_dataframe(macro, "macro")
```

## 4. Data Cleaning

The transaction file intentionally contains realistic data quality issues. We clean them systematically.

Cleaning steps:

1. Convert date columns to datetime.
2. Remove duplicate invoice rows.
3. Standardize text columns.
4. Fill missing financial rates with appropriate values.
5. Recalculate gross margin where needed.
6. Create monthly date variables.

```python
# Convert dates
transactions["invoice_date"] = pd.to_datetime(transactions["invoice_date"])
budget["month"] = pd.to_datetime(budget["month"])
financials["month"] = pd.to_datetime(financials["month"])
macro["month"] = pd.to_datetime(macro["month"])

# Remove duplicates
before = len(transactions)
transactions = transactions.drop_duplicates()
after = len(transactions)

print(f"Duplicate rows removed: {before - after}")

# Clean text columns
text_cols = transactions.select_dtypes(include="object").columns

for col in text_cols:
    transactions[col] = transactions[col].str.strip().str.title()

# Fill missing discount rate with median
transactions["discount_rate"] = transactions["discount_rate"].fillna(
    transactions["discount_rate"].median()
)

# Fill missing days to collect by payment terms group
transactions["days_to_collect"] = transactions.groupby("payment_terms")["days_to_collect"].transform(
    lambda x: x.fillna(x.median())
)

# Recalculate gross margin where missing
transactions["gross_margin_pct"] = transactions["gross_margin_pct"].fillna(
    transactions["gross_profit"] / transactions["net_revenue"]
)

# Create time variables
transactions["month"] = transactions["invoice_date"].dt.to_period("M").dt.to_timestamp()
transactions["year"] = transactions["invoice_date"].dt.year
transactions["quarter"] = transactions["invoice_date"].dt.quarter
transactions["month_name"] = transactions["invoice_date"].dt.month_name()

# Save cleaned data
transactions.to_csv(OUTPUTS_DIR / "cleaned_financial_transactions.csv", index=False)

transactions.head()
```

## 7. Profitability Analysis

Revenue growth is not enough. Finance teams also need to know whether revenue is profitable.

This section compares revenue and gross margin by product line. A product may have high sales but weak margin, which may require pricing changes, supplier negotiations, or discount controls.

```python
profitability = revenue_by_product.copy()

portfolio_gross_margin = transactions["gross_profit"].sum() / transactions["net_revenue"].sum()

profitability["margin_gap_vs_portfolio"] = profitability["gross_margin"] - portfolio_gross_margin
profitability["high_revenue_weak_margin"] = (
    (profitability["revenue_share"] > profitability["revenue_share"].median()) &
    (profitability["gross_margin"] < portfolio_gross_margin)
)

profitability.to_csv(OUTPUTS_DIR / "profitability_by_product_line.csv", index=False)
profitability
```

Discounting can increase sales volume, but it can also weaken margins and increase low-quality demand.

We group transactions into discount bands and compare:

- Return rate

```python
transactions["discount_band"] = pd.cut(
    transactions["discount_rate"],
    bins=[-0.001, 0.05, 0.10, 0.20, 0.30, 1.00],
    labels=["0-5%", "5-10%", "10-20%", "20-30%", "30%+"]
)

discount_analysis = (
    transactions
    .groupby("discount_band", observed=False)
    .agg(
        revenue=("net_revenue", "sum"),
        avg_invoice_revenue=("net_revenue", "mean"),
        gross_margin=("gross_margin_pct", "mean"),
        return_rate=("is_returned", "mean"),
        avg_days_to_collect=("days_to_collect", "mean"),
        invoices=("invoice_id", "nunique")
    )
    .reset_index()
)

discount_analysis.to_csv(OUTPUTS_DIR / "discount_analysis.csv", index=False)
discount_analysis
```

## 9. Budget vs Actual Variance Analysis

Budget variance analysis is central to FP&A work.

We compare actuals to budget by month, region, and product line.

Important variance formulas:

- Revenue variance = Actual revenue minus Budget revenue
- Revenue variance percent = Actual revenue divided by Budget revenue minus 1
- Gross profit variance = Actual gross profit minus Budget gross profit

Positive variance is favorable for revenue and profit. Positive COGS variance is usually unfavorable if actual costs exceed budget.

```python
actual_monthly_detail = (
    transactions
    .groupby(["month", "region", "product_line"], as_index=False)
    .agg(
        actual_revenue=("net_revenue", "sum"),
        actual_cogs=("cogs", "sum"),
        actual_gross_profit=("gross_profit", "sum"),
        actual_units=("units", "sum")
    )
)

variance = actual_monthly_detail.merge(
    budget,
    on=["month", "region", "product_line"],
    how="left"
)

variance["revenue_variance"] = variance["actual_revenue"] - variance["budget_revenue"]
variance["revenue_variance_pct"] = variance["actual_revenue"] / variance["budget_revenue"] - 1

variance["cogs_variance"] = variance["actual_cogs"] - variance["budget_cogs"]
variance["gross_profit_variance"] = variance["actual_gross_profit"] - variance["budget_gross_profit"]
variance["gross_profit_variance_pct"] = variance["actual_gross_profit"] / variance["budget_gross_profit"] - 1

variance["unit_variance"] = variance["actual_units"] - variance["budget_units"]

variance.to_csv(OUTPUTS_DIR / "budget_vs_actual_variance.csv", index=False)

variance.head()
```

```python
variance_summary = (
    variance
    .groupby("product_line", as_index=False)
    .agg(
        actual_revenue=("actual_revenue", "sum"),
        budget_revenue=("budget_revenue", "sum"),
        revenue_variance=("revenue_variance", "sum"),
        actual_gross_profit=("actual_gross_profit", "sum"),
        budget_gross_profit=("budget_gross_profit", "sum"),
        gross_profit_variance=("gross_profit_variance", "sum")
    )
)

variance_summary["revenue_variance_pct"] = (
    variance_summary["actual_revenue"] / variance_summary["budget_revenue"] - 1
)

variance_summary["gross_profit_variance_pct"] = (
    variance_summary["actual_gross_profit"] / variance_summary["budget_gross_profit"] - 1
)

variance_summary = variance_summary.sort_values("revenue_variance", ascending=False)

variance_summary.to_csv(OUTPUTS_DIR / "budget_variance_summary_by_product.csv", index=False)
variance_summary
```

```python
plt.figure(figsize=(11, 6))
sns.barplot(
    data=variance_summary,
    x="product_line",
    y="revenue_variance"
)
plt.axhline(0, color="black", linewidth=1)
plt.title("Revenue Budget Variance by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Actual Revenue minus Budget Revenue")
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig(FIGURES_DIR / "budget_variance_by_product_line.png", dpi=300)
plt.show()
```

## 11. Working Capital and Cash Flow Analysis

Revenue and profit do not automatically mean cash generation.

This section calculates:

- Days Sales Outstanding
- Inventory days
- Payables days
- Cash conversion cycle
- Free cash flow margin

A rising cash conversion cycle means cash is tied up longer in receivables and inventory.

```python
working_capital = financials.copy()

working_capital["days_in_month"] = working_capital["month"].dt.days_in_month

working_capital["dso"] = (
    working_capital["accounts_receivable"] / working_capital["revenue"]
) * working_capital["days_in_month"]

working_capital["inventory_days"] = (
    working_capital["inventory"] / working_capital["cogs"]
) * working_capital["days_in_month"]

working_capital["payables_days"] = (
    working_capital["accounts_payable"] / working_capital["cogs"]
) * working_capital["days_in_month"]

working_capital["cash_conversion_cycle"] = (
    working_capital["dso"]
    + working_capital["inventory_days"]
    - working_capital["payables_days"]
)

working_capital["free_cash_flow_margin"] = (
    working_capital["free_cash_flow"] / working_capital["revenue"]
)

working_capital.to_csv(OUTPUTS_DIR / "working_capital_analysis.csv", index=False)

working_capital[[
    "month", "dso", "inventory_days", "payables_days",
    "cash_conversion_cycle", "free_cash_flow_margin"
]].tail()
```

```python
plt.figure(figsize=(13, 6))
sns.lineplot(data=working_capital, x="month", y="cash_conversion_cycle")
plt.title("Cash Conversion Cycle Over Time")
plt.xlabel("Month")
plt.ylabel("Days")
plt.tight_layout()
plt.savefig(FIGURES_DIR / "cash_conversion_cycle.png", dpi=300)
plt.show()
```

## 13. Forecasting Dataset Preparation

We now build a modeling dataset.

Forecasting features include:

- Month number
- Time trend
- Lagged revenue
- Rolling average revenue
- Macro indicators
- Seasonal dummy variables

The target variable is monthly revenue.

```python
forecast_df = financials[["month", "revenue"]].merge(macro, on="month", how="left")
forecast_df = forecast_df.sort_values("month")

forecast_df["time_index"] = np.arange(len(forecast_df))
forecast_df["month_num"] = forecast_df["month"].dt.month
forecast_df["year"] = forecast_df["month"].dt.year

forecast_df["revenue_lag_1"] = forecast_df["revenue"].shift(1)
forecast_df["revenue_lag_3"] = forecast_df["revenue"].shift(3)
forecast_df["revenue_lag_12"] = forecast_df["revenue"].shift(12)
forecast_df["revenue_roll_3"] = forecast_df["revenue"].shift(1).rolling(3).mean()
forecast_df["revenue_roll_6"] = forecast_df["revenue"].shift(1).rolling(6).mean()

forecast_df = pd.get_dummies(forecast_df, columns=["month_num"], drop_first=True)

model_df = forecast_df.dropna().copy()

model_df.head()
```

## 14. Train-Test Split

Because this is time series data, we do not randomly split the data. We train on earlier months and test on the most recent 12 months.

This mimics a real forecasting environment where the future is unknown.

```python
test_months = 12

train = model_df.iloc[:-test_months].copy()
test = model_df.iloc[-test_months:].copy()

target = "revenue"
drop_cols = ["month", target]

X_train = train.drop(columns=drop_cols)
y_train = train[target]

X_test = test.drop(columns=drop_cols)
y_test = test[target]

print("Training period:", train["month"].min(), "to", train["month"].max())
print("Testing period:", test["month"].min(), "to", test["month"].max())
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

## 15. Forecast Evaluation Functions

We evaluate forecast accuracy using:

- MAE: Mean Absolute Error
- RMSE: Root Mean Squared Error
- MAPE: Mean Absolute Percentage Error

MAPE is especially useful for finance audiences because it expresses error as a percentage.

```python
def mape(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true))

def evaluate_forecast(model_name, y_true, y_pred):
    return {
        "model": model_name,
        "mae": mean_absolute_error(y_true, y_pred),
        "rmse": np.sqrt(mean_squared_error(y_true, y_pred)),
        "mape": mape(y_true, y_pred)
    }
```

## 16. Model 1: Seasonal Naive Forecast

The seasonal naive model is a simple but useful benchmark.

It predicts this month using the value from the same month last year.

A sophisticated model should beat this simple baseline.

```python
test_baseline = test[["month", "revenue", "revenue_lag_12"]].copy()
test_baseline["seasonal_naive_forecast"] = test_baseline["revenue_lag_12"]

baseline_results = evaluate_forecast(
    "Seasonal Naive",
    test_baseline["revenue"],
    test_baseline["seasonal_naive_forecast"]
)

baseline_results
```

## 17. Model 2: Exponential Smoothing

Exponential smoothing is a standard time series forecasting method. It can capture trend and seasonality.

This is highly relevant for finance teams that forecast monthly revenue, sales, or expenses.

```python
ts = financials.set_index("month")["revenue"].asfreq("MS")

train_ts = ts.iloc[:-test_months]
test_ts = ts.iloc[-test_months:]

ets_model = ExponentialSmoothing(
    train_ts,
    trend="add",
    seasonal="add",
    seasonal_periods=12
).fit(optimized=True)

ets_forecast = ets_model.forecast(test_months)

ets_results = evaluate_forecast(
    "Exponential Smoothing",
    test_ts.values,
    ets_forecast.values
)

ets_results
```

## 18. Model 3: Machine Learning Forecast

We use a Random Forest model with lagged revenue, rolling averages, trend, and macroeconomic variables.

This is not a pure time series model, but it is useful when finance teams want to include external drivers such as interest rates, business confidence, inflation, or market indicators.

```python
rf = RandomForestRegressor(
    n_estimators=500,
    max_depth=8,
    random_state=42
)

rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)

rf_results = evaluate_forecast(
    "Random Forest",
    y_test,
    rf_pred
)

rf_results
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)

lr_results = evaluate_forecast(
    "Linear Regression",
    y_test,
    lr_pred
)

lr_results
```

```python
model_results = pd.DataFrame([
    baseline_results,
    ets_results,
    rf_results,
    lr_results
]).sort_values("mape")

model_results.to_csv(OUTPUTS_DIR / "forecast_model_results.csv", index=False)
model_results
```

## 22. Scenario Analysis

Finance teams often need scenario analysis, not just one forecast.

We create three scenarios:

1. Downside case
2. Base case
3. Upside case

Each scenario adjusts revenue growth and margin assumptions.

```python
last_actual_revenue = financials["revenue"].iloc[-1]
last_gross_margin = financials["gross_margin"].iloc[-1]
last_ebitda_margin = financials["ebitda_margin"].iloc[-1]

scenario_assumptions = pd.DataFrame({
    "scenario": ["Downside", "Base", "Upside"],
    "monthly_revenue_growth": [-0.01, 0.01, 0.025],
    "gross_margin_change": [-0.015, 0.000, 0.012],
    "ebitda_margin_change": [-0.020, 0.000, 0.018]
})

scenario_rows = []

for _, row in scenario_assumptions.iterrows():
    revenue = last_actual_revenue
    gross_margin = last_gross_margin + row["gross_margin_change"]
    ebitda_margin = last_ebitda_margin + row["ebitda_margin_change"]

    for m in range(1, 7):
        revenue = revenue * (1 + row["monthly_revenue_growth"])
        gross_profit = revenue * gross_margin
        ebitda = revenue * ebitda_margin

        scenario_rows.append({
            "scenario": row["scenario"],
            "period": m,
            "forecast_revenue": revenue,
            "forecast_gross_profit": gross_profit,
            "forecast_ebitda": ebitda,
            "assumed_gross_margin": gross_margin,
            "assumed_ebitda_margin": ebitda_margin
        })

scenario_df = pd.DataFrame(scenario_rows)

scenario_summary = (
    scenario_df
    .groupby("scenario", as_index=False)
    .agg(
        six_month_revenue=("forecast_revenue", "sum"),
        six_month_gross_profit=("forecast_gross_profit", "sum"),
        six_month_ebitda=("forecast_ebitda", "sum")
    )
)

scenario_df.to_csv(OUTPUTS_DIR / "scenario_analysis_detail.csv", index=False)
scenario_summary.to_csv(OUTPUTS_DIR / "scenario_analysis_summary.csv", index=False)

scenario_summary
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=scenario_df,
    x="period",
    y="forecast_revenue",
    hue="scenario",
    marker="o"
)
plt.title("Revenue Scenario Analysis")
plt.xlabel("Forecast Month")
plt.ylabel("Forecast Revenue")
plt.tight_layout()
plt.savefig(FIGURES_DIR / "revenue_scenario_analysis.png", dpi=300)
plt.show()
```
